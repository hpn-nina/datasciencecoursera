```python
import numpy as np
```

```python
# Read data from file Data\GIS\Population\population_HCMC\population_shapefile\Population_Ward_Level.shp
import geopandas as gpd
shapefile = gpd.read_file("./CSL_HCMC/Data/GIS/Population/population_HCMC/population_shapefile/Population_Ward_Level.shp")
print(shapefile)
```

```python
# Highest lowest
    ## Population
    ## Density
    ## Area
    ## Rate of leveling up
    ## Rate of changing
```

```python
# Population
# Highest population in 2019
# Phuong co dan so cao nhat
highestPop = shapefile[shapefile.Pop_2019 >= max(shapefile['Pop_2019'])]
highestPop

```

```python
# Lowest population in 2019
# Phuong co dan so thap nhat
lowestPop = shapefile[shapefile.Pop_2019 <= min(shapefile['Pop_2019'])]
lowestPop
```

```python
# Density
# Highest density 
# Mat do dan so cao nhat
highestDen = shapefile[shapefile.Den_2019 >= max(shapefile['Den_2019'])]
highestDen
```

```python
# Lowest density 
# Mat do dan so thap nhat
lowestDen = shapefile[shapefile.Den_2019 <= min(shapefile['Den_2019'])]
lowestDen
```

```python
# Area
# Biggest area 
# Phuong co dien tich lon nhat
biggestArea = shapefile[shapefile.Shape_Area >= max(shapefile['Shape_Area'])]
biggestArea
```

```python
# Smallest area 
# Phuong co dien tich nho nhat
smallestArea = shapefile[shapefile.Shape_Area <= min(shapefile['Shape_Area'])]
smallestArea
```

```python
# Bien dong dan so
# Change in population
# Highest and lowest
shapefile['speed'] = shapefile['Pop_2019'] - shapefile['Pop_2009']
highestSpeed = shapefile[shapefile.speed >= max(shapefile['speed'])]
lowestSpeed = shapefile[shapefile.speed <= min(shapefile['speed'])]

print(highestSpeed, lowestSpeed)
```

```python
# Toc do tang truong
# Calculate the percentage

shapefile['percentage'] = shapefile['Pop_2019'] / shapefile['Pop_2009']
highestPercentage = shapefile[shapefile.percentage >= max(shapefile['percentage'])]
lowestPercentage = shapefile[shapefile.percentage <= min(shapefile['percentage'])]

print(highestPercentage, lowestPercentage)
```
